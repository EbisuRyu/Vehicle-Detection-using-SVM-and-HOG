```python
import cv2
from script.feature_source import FeatureExtracter
from script.helpers import show_images
import matplotlib.pyplot as plt 
sourcer_params = {
  'color_model': 'gray',                # hls, hsv, yuv, ycrcb 
  'spatial_size': (200, 200),            # (16, 16), (32, 32), (64, 64)
  'orientations': 9,        # 6 - 12
  'pixels_per_cell': 8,               # 8, 16
  'cells_per_block': 2,                # 1, 2
  'transform_sqrt': True,
  'block_norm': 'L2',
  'hog_visualize': True
}
sourcer = FeatureExtracter(**sourcer_params)
```

```python
import cv2
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt

# c hnh nh xm t tp
gray_image = cv2.imread('./dataset/traffic_sign_board/images/road100.png', cv2.IMREAD_GRAYSCALE)
#gray_image = cv2.imread('./dataset/vehicles/GTI_Right/image0001.png', cv2.IMREAD_GRAYSCALE)

gray_image = cv2.resize(gray_image, (200, 200))
# Trch xut HOG ca hnh nh xm
features, hog_image = hog(gray_image, orientations=12, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True)

# Cn chnh  sng ca hnh nh HOG
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))

# Hin th hnh nh HOG
plt.imshow(hog_image_rescaled, cmap='gray')
plt.axis('off')  # n trc
plt.show()
```

```python
for i in range(83, 103):
    start_frame = cv2.imread(f'./dataset/traffic_sign_board/images/road{i}.png', cv2.IMREAD_GRAYSCALE)
    features = sourcer.features(start_frame)
    print("feature shape:", features.shape)
    a_img = sourcer.visualize()
    show_images([a_img], per_row = 2, per_col = 1, W = 10, H = 2)
```
